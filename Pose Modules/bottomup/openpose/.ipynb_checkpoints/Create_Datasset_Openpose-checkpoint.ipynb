```python
def get_features(video_dire):
    extracted_features=[]
    avg_confs=[]
    missing_joints=[]
    video_reader=cv.VideoCapture(video_dire)
    total_frames=int(video_reader.get(cv.CAP_PROP_FRAME_COUNT))
    skip_frames=max(int(total_frames/seq_len),1)
    for frame_counter in range(total_frames):
        video_reader.set(cv.CAP_PROP_POS_FRAMES,frame_counter*skip_frames)
        sucess,frame=video_reader.read()
        if not sucess:
            break
        frame=cv2.resize(frame,(w,h))    
        feature,df=get_features_df(frame)
        extracted_features.append(feature)
        avg_confs.append(np.average(df['confidence']))
        missing_joints.append(len(df[df['confidence']==0]['confidence']))
        if len(extracted_features)==seq_len:
            break
    while (len(extracted_features)<seq_len):
        extracted_features.append(np.zeros(36,np.float32))    
    return np.array(extracted_features),np.average(avg_confs),np.average(missing_joints)
```

```python
#for AIHUB
def create_dataset():
    labels=[]
    features=[]
    dataset_avg_conf=[]
    dataset_missing_points=[]
    for idx,className in enumerate(CLASSES_LIST):
        print(className)
        video_dires_list=glob(os.path.join(data_set_dire,className)+"/*")
        
        random.shuffle(video_dires_list)
        if len(video_dires_list)>=500:
            video_dires_list=video_dires_list[:499]
        
        for num_videes,video_dire in enumerate(video_dires_list):
            if num_videes %10==0:
                print("Num of Videos Processed : {}".format((idx+1)*num_videes))
            video_dire=video_dire.replace("\\","/")
            
            extracted_features,avg_video_conf,avg_video_missing_point= get_features(video_dire)
            dataset_avg_conf.append(avg_video_conf)
            dataset_missing_points.append(avg_video_missing_point)
            
            
            features.append(extracted_features)
            labels.append(idx)
                
    return features,labels,dataset_avg_conf,dataset_missing_points
```

```python
features,labels,dataset_avg_conf,dataset_missing_points=create_dataset()
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.33, random_state=42)
saved_dire=os.path.join(train_data_dire,'openpose','AIHUB')
if not os.path.exists(saved_dire):
    os.makedirs(saved_dire)
np.save("{}/X_train".format(saved_dire),X_train)

np.save("{}/X_test".format(saved_dire),X_test)

np.save("{}/y_train".format(saved_dire),y_train)

np.save("{}/y_test".format(saved_dire),y_test)
np.save("{}/avg_conf".format(saved_dire), dataset_avg_conf)
np.save("{}/missiing_points".format(saved_dire), dataset_missing_points)

np.save("{}/features".format(saved_dire),features)

np.save("{}/labels".format(saved_dire),labels)
```

```python

## For eldersim
def create_dataset():
    labels=[]
    features=[]
    dataset_avg_conf=[]
    dataset_missing_points=[]
    for idx,className in enumerate(CLASSES_LIST):
        print(className)
        
        if className=='walk':
            video_dires_list=glob(os.path.join(data_set_dire,className)+"/*.avi")
        else:
            video_dires_list=glob(os.path.join(data_set_dire,className)+"/**/*.avi")
        
        random.shuffle(video_dires_list)
        if len(video_dires_list)>=1600:
            video_dires_list=video_dires_list[:1600]
        for num_videes,video_dire in enumerate(video_dires_list):
            if num_videes %10==0:
                print("Num of Videos Processed : {}".format((idx+1)*num_videes))
            video_dire=video_dire.replace("\\","/")
            
            extracted_features,avg_video_conf,avg_video_missing_point= get_features(video_dire)
            dataset_avg_conf.append(avg_video_conf)
            dataset_missing_points.append(avg_video_missing_point)
            
            
            features.append(extracted_features)
            labels.append(idx)
                
    return features,labels,dataset_avg_conf,dataset_missing_points
```

```python
features,labels,dataset_avg_conf,dataset_missing_points=create_dataset()
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.33, random_state=42)
saved_dire=os.path.join(train_data_dire,'openpose','Eldersim')
if not os.path.exists(saved_dire):
    os.makedirs(saved_dire)
np.save("{}/X_train".format(saved_dire),X_train)

np.save("{}/X_test".format(saved_dire),X_test)

np.save("{}/y_train".format(saved_dire),y_train)

np.save("{}/y_test".format(saved_dire),y_test)
np.save("{}/avg_conf".format(saved_dire), dataset_avg_conf)
np.save("{}/missiing_points".format(saved_dire), dataset_missing_points)

np.save("{}/features".format(saved_dire),features)

np.save("{}/labels".format(saved_dire),labels)
```
